## Least Square Method

```python
import numpy as np
import matplotlib.pyplot as plt

def leastsq(x,y,n):
    xi=0
    yi=0
    xi2=0
    xiyi=0
    m=0
    c=0
    for i in range(len(x)):
        xi+=x[i]
        yi+=y[i]
        xi2+=(x[i])**2
        xiyi+=x[i]*y[i]
    #Calculate m and c
    m=(n*xiyi-xi*yi)/(n*xi2-(xi**2))
    c=(yi-m*xi)/n
    return m,c
```

```python
#Vectors
np.random.seed(2)
x=np.random.rand(10)
x.sort()
y=np.random.rand(10)
y.sort()
```

```python
#My least squares
mm,cc=leastsq(x,y,len(x))
print(f"Slope: {mm}, Intercept: {cc}")
```

```python
#Numpy least squares
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

### Testing Code

```python
#Generated points
mi=10.5
ci=20
x2=np.arange(0,1,.01)
x2.sort()
ran=np.random.uniform(-1,1,size=len(x2))
y2=mi*x2+ci+ran

print(f"Real slope: {mi}, Real intercept: {ci}")
```

```python
#My code
mi2,ci2=leastsq(x2,y2,len(x2))
print(f"Calculated slope: {mi2}, Calculated intercept: {ci2}")
```

```python
#Finding new y and residuals
y_calc=mi2*x2+ci2

residual=y2-y_calc
print("Maximum residual: ",max(abs(residual)))
print("Minimum residual: ",min(abs(residual)))
print(residual)
```

```python
#Plotting
plt.figure()
plt.scatter(x2,y2,label='x,y')
plt.plot(x2,y_calc,label='line of best fit')
plt.legend()
plt.show()
```
