## Plot Monte Carlo Pi

##### Iterate the number of samples for your Monte Carlo Pi approximation from 10 to 10000, in steps of 20, and plot the error (absolute value of your calculated value of Pi minus the actual value of Pi encoded in Python, to 15 digits past the decimal point)

```python
import math
import random
def apPi(numSample):
    counter=0
    for s in range(numSample):
        x=random.uniform(-1,1)
        y=random.uniform(-1,1)
       
        if(x*x+y*y<1):
            counter=counter+1
    return 4*counter/numSample
```

```python
import numpy as np
import matplotlib.pyplot as plt
error=[]
er=0
mont=0
i=10
for i in range(1,10000):
    mont=apPi(i)
    er=abs(np.pi-mont)
    error.append(er)
    i=i+20
    
plt.figure()
plt.plot(error,color='gold')
plt.xlabel('Trial')
plt.ylabel('Error')
plt.title('Absolute Error')

plt.show()
    
```
