## Taylor Series, Truncation Error

```python
import numpy as np
import math
import matplotlib.pyplot as plt

def sinApp(x,n):
    sin=0
    for i in range(n):
        side=(-1)**i
        top=x**(2*i+1)
        bottom=math.factorial(2*i+1)
        sin=sin+((side*top)/bottom)
    return sin
```

```python
sin1=[]
error1=[]
x=[]
for i in np.arange(0.,10.,.1):
    sin=sinApp(i,1)
    sin1.append(sin)
    x.append(i)
    
    if i==0:
        error1.append(0)
    else:
        error=abs(np.sin(i)-sin)
        error1.append(error)
```

```python
sin2=[]
error2=[]
for i in np.arange(0.,10.,.1):
    sin=sinApp(i,2)
    sin2.append(sin)
    
    if i==0:
        error2.append(0)
    else:
        error=abs(np.sin(i)-sin)
        error2.append(error)
```

```python
sin3=[]
error3=[]
for i in np.arange(0.,10.,.1):
    sin=sinApp(i,3)
    sin3.append(sin)
    
    if i==0:
        error3.append(0)
    else:
        error=abs(np.sin(i)-sin)
        error3.append(error)
```

```python
sin4=[]
error4=[]
for i in np.arange(0.,10.,.1):
    sin=sinApp(i,4)
    sin4.append(sin)
    
    if i==0:
        error4.append(0)
    else:
        error=abs(np.sin(i)-sin)
        error4.append(error)
        
```

```python
import matplotlib.pyplot as plt

plt.subplot(2,2,1)
plt.plot(x,error1,color='purple')
plt.title("Error Using 1 Term")
plt.xlabel("x-value")
plt.ylabel("Truncation Error")

plt.subplot(2,2,2)
plt.plot(x,error2,'b')
plt.title("Error Using 2 Terms")
plt.xlabel("x-value")
plt.ylabel("Truncation Error")

plt.subplot(2,2,3)
plt.plot(x,error3,'g')
plt.title("Error Using 3 Terms")
plt.xlabel("x-value")
plt.ylabel("Truncation Error")

plt.subplot(2,2,4)
plt.plot(x,error4,'r')
plt.title("Error Using 4 Terms")
plt.xlabel("x-value")
plt.ylabel("Truncation Error")

plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.title('Comparing Actual Values of Sin With Calculated Value')
plt.xlabel('x-value in sin(x)')
plt.ylabel('Truncation Error')
plt.plot(x,error1,'m',label='One Term')
plt.plot(x,error2,'b',label='Two Terms')
plt.plot(x,error3,'g',label='Three Terms')
plt.plot(x,error4,'r',label='Four Terms')
plt.legend()
plt.show()
```

##### Discuss how the truncation error behaves as you include more terms. Does the error decrease uniformly for all values of x? Does the truncation error behave differently for small vs. large values of x (try feeding larger values of x)? (a few sentences is fine)

### As more terms are included, the truncation error decreases for more x values while the error for the larger x values increases exponentially. This is shown when comparing the purple(one term) and the red(four terms) lines. The error for the purple line does not change much when comparing smaller x values to the larger ones. It almost looks like a horizontal line compared to the red one looking like an exponential function.
