# Monte Carlo Pi

```python
import math
import random
def apPi(numSample):
    counter=0
    for s in range(numSample):
        x=random.uniform(-1,1)
        y=random.uniform(-1,1)
       
        if(x*x+y*y<1):
            counter=counter+1
    return 4*counter/numSample
```

```python
x=apPi(25)
print(x)
```
