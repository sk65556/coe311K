##### Create a line graph using matplotlib that shows monthly sales data for the year 2024. The x-axis should display months (January through December) and the y-axis should show sales amounts ranging from $10,000 to $50,000. Use random or sample data, add a title 'Monthly Sales Report 2024', label both axes appropriately, include a grid for easier reading, and use a blue line with circular markers.

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.random.randint(10000,50000,12)
x=np.array(["January","February", "March", "April","May","June","July", "August", "September", "October", "November", "December"])

plt.plot(x,y,'m-',marker='o')
plt.title('Monthly Sales Report 2024')
plt.xlabel('Months')
plt.ylabel('Sale Amounts')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig('graph.png')
plt.show()
```
