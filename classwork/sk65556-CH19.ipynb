```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
hs = [0.1,0.01,0.001]  # step size
x_start, x_end = 0, 30
exact=[]

def f(x, y):
    return -1000*y+3000-2000*np.exp(-x)
def exactsol(x):
    return 3-0.998*np.exp(-1000*x)-2.002*np.exp(-x)

for h in hs:
    num_steps = int((x_end - x_start) / h)
    x_values=np.linspace(x_start, x_end, num_steps + 1)
    y_values=np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 0  # y(0) = 0


    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
    
    exact=exactsol(x_values)
    
    # Plot the results
    plt.plot(x_values, y_values, label=f"h={h}")
    plt.plot(x_values,exact,label=f"Exact at h={h}")
plt.title('Approximation of dy/dx= -1000y + 3000 - 2000e^-x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()
plt.show()
```
