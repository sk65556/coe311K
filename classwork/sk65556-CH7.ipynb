## Code Challenge A

```python
import numpy as np
import time
def jacobi(A, b, x0, tol, max_iterations):
    start=time.perf_counter()
    n = len(b)
    x = x0.copy()
    #print(x)
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            timer=time.perf_counter()-start
            return x_new, k,timer
        x = x_new
        #print(x)
    
    timer=time.perf_counter()-start
    return x, max_iterations,timer

def gauss_seidel(A, b, x0, tol, max_iterations):
    start=time.perf_counter()
    n = len(b)
    x = x0.copy()
    #print(x)
    
    for k in range(max_iterations):
        x_new = x.copy()
        
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
            # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            timer=time.perf_counter()-start
            return x_new, k,timer
        x = x_new
        #print(x)
    
    timer=time.perf_counter()-start
    return x, max_iterations,timer
```

```python
A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)
```

```python
tol=1e-6
x0=np.zeros_like(b)
maxit=100
sol,ite,timer=jacobi(A,b,x0,tol,maxit)
print('The solution using Jacobi is:',sol,'It took ',ite,'iterations', 'In ',timer,'of a second')

solG,iteG,timerG=gauss_seidel(A,b,x0,tol,maxit)
print('The solution using Gauss-Seidel is:',solG,'It took ',iteG,'iterations', 'In ',timerG,'of a second')
```

## Coding Challenge Part D


```python
def diagdom(A):
    for i in range(len(A)):
        sums=0
        for j in range(i):
            while (i!=j):
                sums+=abs(A[i,j])
                if (abs(A[i,i])>=sums):
                    return True
                else:
                    return False
    
```

```python
A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
```
