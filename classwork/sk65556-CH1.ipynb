```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
rng=default_rng(2)

x=np.linspace(1,2*np.pi,30)

plt.subplot(2,2,1)
plt.plot(x,np.cos(x),color='purple')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('graph of cos')

plt.subplot(2,2,2)
plt.scatter(x**2,np.cos(x)**2,color='orange')
plt.xlabel('x')
plt.ylabel('y')


z=x**2+2*x-4
r=3*x**2+x+2
plt.subplot(2,2,3)
plt.plot(x,z,color='gold',label='Maxine')
plt.plot(x,r,color='silver',label='Jessica')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Money in Dollars')
plt.title("Jessica and Maxine's Income")


a=rng.random(1000)
plt.subplot(2,2,4)
plt.hist(a,10,color='skyblue',edgecolor='blue')
plt.xlabel('Year')
plt.ylabel('Waterfall in Inches')
plt.title('Waterfall Per Region Per Year')

plt.show
```
