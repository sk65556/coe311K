```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and the domain
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi  # Interval
N = 10  # Number of sample points

# Create a uniform grid
x = np.linspace(a, b, N)
y = f(x)

# Plot the sampled points
plt.plot(x, y, 'o-', label='Sampled Points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Uniform Sampling of f(x) = sin(x)')
plt.legend()
plt.show()
```

```python
# Step size
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
df_dx = (f(x + h) - f(x - h)) / (2 * h)

# Plot the approximation and the true derivative
plt.plot(x, np.cos(x), label='Exact Derivative')
plt.plot(x, df_dx, 'o--', label='Forward Difference Approximation')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Approximating Derivative of f(x) = sin(x) using Central Difference")
plt.legend()
plt.show()
```

```python
# Step size
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
df_dxc = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
df_dxf = (f(x + h) - f(x)) / h

# Backward Difference Approximation
df_dxb = (f(x) - f(x - h)) / h


# Plot the approximation and the true derivative
## Do a canvas of 4 subplots or 4 graphs

plt.figure()

plt.subplot(2,2,1)
plt.plot(x, np.cos(x),color='blue')
plt.title('Exact Derivative')
plt.xlabel('x')
plt.ylabel("f'(x)")

plt.subplot(2,2,2)
plt.plot(x, df_dxc, 'o--',color='green')
plt.title("Central Difference Approximation")
plt.xlabel('x')
plt.ylabel("f'(x)")

plt.subplot(2,2,3)
plt.plot(x, df_dxf, 'o--',color='purple')
plt.title("Forward Difference Approximation")
plt.xlabel('x')
plt.ylabel("f'(x)")

plt.subplot(2,2,4)
plt.plot(x, df_dxb, 'o--',color='orange')
plt.title("Backward Difference Approximation")
plt.xlabel('x')
plt.ylabel("f'(x)")

plt.tight_layout()
plt.show()

```
