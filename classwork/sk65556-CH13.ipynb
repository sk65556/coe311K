```python
import numpy as np
import time
import matplotlib.pyplot as plt

np.random.seed(13)

#Creating the Matrices
A=np.random.random((3,3))
B=np.random.random((3,3))

C=np.random.random((6,6))
D=np.random.random((6,6))

E=np.random.random((12,12))
F=np.random.random((12,12))

G=np.random.random((24,24))
H=np.random.random((24,24))

I=np.random.random((48,48))
J=np.random.random((48,48))

K=np.random.random((96,96))
L=np.random.random((96,96))

M=np.random.random((192,192))
N=np.random.random((192,192))

O=np.random.random((384,384))
P=np.random.random((384,384))
```

```python
#Naive Matrix Multiplication
def naive_matrix_multiplication(A, B):
    start=time.perf_counter()
    n = len(A)
    C = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    timer=time.perf_counter()-start
    return timer
```

```python
#Python Matrix Multiplication Timer
def matmul_timer(A,B):
    start=time.perf_counter()
    x=np.dot(A,B)
    end=time.perf_counter()
    timer=end-start
    return timer
```

```python
#Times
naive_time=[]
python_time=[]

timen=naive_matrix_multiplication(A,B)
naive_time.append(timen)
timep=matmul_timer(A,B)
python_time.append(timep)

timen=naive_matrix_multiplication(C,D)
naive_time.append(timen)
timep=matmul_timer(C,D)
python_time.append(timep)

timen=naive_matrix_multiplication(E,F)
naive_time.append(timen)
timep=matmul_timer(E,F)
python_time.append(timep)

timen=naive_matrix_multiplication(G,H)
naive_time.append(timen)
timep=matmul_timer(G,H)
python_time.append(timep)

timen=naive_matrix_multiplication(I,J)
naive_time.append(timen)
timep=matmul_timer(I,J)
python_time.append(timep)

timen=naive_matrix_multiplication(K,L)
naive_time.append(timen)
timep=matmul_timer(K,L)
python_time.append(timep)

timen=naive_matrix_multiplication(M,N)
naive_time.append(timen)
timep=matmul_timer(M,N)
python_time.append(timep)

timen=naive_matrix_multiplication(O,P)
naive_time.append(timen)
timep=matmul_timer(O,P)
python_time.append(timep)

print(naive_time)
print(python_time)
```

```python
#Plot
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8]
plt.figure()
plt.plot(x,naive_time,color='purple',label='Naive',marker='o')
plt.plot(x,python_time,color='blue',label='Optimized',marker='o')
plt.ylabel('Runtime')
plt.xlabel('Matrix Number')
plt.yscale('log')
plt.legend()
plt.show()
```
