## Least Squares

### Part 1

```python
import numpy as np
# Data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

print('X:')
print(X)
print('Y:')
print(Y)
print('A')
print(A)
```

```python
# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(X,Y,color="#512da8",label='X,Y')
x=np.linspace(min(X),max(X),5)
y=m*X+c
plt.plot(X,y,label='y=0.95x+0.59',color='#01153E')
plt.legend()
plt.show()
```

### Part 2

```python
from numpy.random import default_rng
rng=default_rng(2)

#Randomizing the vectors
X=rng.random(5)
Y=rng.random(5)
Y.sort()
X.sort()

A = np.vstack([X, np.ones(len(X))]).T

```

```python
#Solving using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
#Plotting the data points
plt.figure()
plt.scatter(X,Y,color="#512da8",label='X,Y')
#Plotting the line of best fit
y=m*X+c
plt.plot(X,y,label='y=mx+c',color='#01153E')
plt.legend()
plt.show()
```

## Part 3

```python
#Original Equation
x=np.arange(0,1,.01)
x=np.sort(x)
ran=np.random.uniform(-1,1,size=len(x))
y=0.5*x+0.15+ran
#plt.scatter(x,y)
```

```python
#Solving using least squares
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
#Finding new y and residuals
y_calc=m*x+c

residual=y-y_calc
print("Maximum residual: ",max(abs(residual)))
print("Minimum residual: ",min(abs(residual)))
print(residual)
```

```python
#Plotting
plt.figure()
plt.scatter(x,y,label='x,y')
plt.plot(x,y_calc,label='line of best fit')
plt.legend()
plt.show()
```
