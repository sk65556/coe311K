```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

def apPi(numSample):
    counter=0
    for s in range(0,numSample):
        x=random.uniform(-1,1)
        y=random.uniform(-1,1)
       
        if(x*x+y*y<1):
            counter=counter+1
    return 4*counter/numSample

import numpy as np
import math
import matplotlib.pyplot as plt
abE=0
relE=0
ab=[]
rel=[]
i=0
for i in range (0,100):
    s=apPi(1000)
    abE=abs(np.pi-s)
    relE=abs((np.pi-s)/np.pi)
    ab.append(abE)
    rel.append(relE)
    


plt.figure()
plt.subplot(2,1,1)
plt.plot(ab)
plt.xlabel('Trial')
plt.ylabel('Error')
plt.title('Absolute Error')

plt.subplot(2,1,2)
plt.plot(rel)
plt.xlabel('Trial')
plt.ylabel('Error')
plt.title('Relative Error')

plt.tight_layout
plt.show()
```
