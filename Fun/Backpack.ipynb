### Create a simple inventory management system for a text-based adventure game.

### Your task:
##### Create a class called Inventory that can:
##### Add items to the inventory
##### Remove items from the inventory
##### Display all items
##### Check if a specific item exists
##### Have a maximum capacity (e.g., 10 items)

### Requirements:
##### Create an __init__ method that sets up an empty inventory with a max capacity
##### add_item(item_name) - adds an item if there's space, otherwise prints "Inventory full!"
##### remove_item(item_name) - removes an item if it exists, otherwise prints "Item not found!"
##### show_inventory() - displays all items in a nice format
##### has_item(item_name) - returns True if the item exists, False otherwise

```python
class Inventory:
    def __init__(self,max_capacity):
        self.max_capacity=max_capacity
        self.items=[]
    def add_item(self,item_name):
        if len(self.items)<self.max_capacity:
            self.items.append(item_name)
            print(f"{item_name} was successfully added")
        else:
            print(f"{item_name} not added due to lack of space")
    def remove_item(self,item_name):
        if item_name in self.items:
            self.items.remove(item_name)
            print(f"Successfully removed {item_name}")
        else:
            print("item not found")
    def show_inventory(self):
        print(self.items)
    def has_item(self,item_name):
        if item_name in self.items:
            print("Yes you have ",item_name," in your inventory")
        else:
            print("You do not have ", item_name, " in your inventory")
            
```

```python
backpack=Inventory(10)
backpack.items = ["sword", "water bottle", "extra clothes", "dagger", "compass", "waterproof pouch", "whistle", "flare"]


action=input("What do you want to do? (add/remove/show/search/done): ")
while action != "done":
    if action=="add":
        item=input("What would you like to add? ")
        backpack.add_item(item)
    elif action=="remove":
        item=input("What would you like to remove? ")
        backpack.remove_item(item)
    elif action=="show":
        backpack.show_inventory()
    elif action=="search":
        item=input("What are you looking for? ")
        backpack.has_item(item)
    else:
        print("Not a valid action try again ")
    action = input("\nWhat do you want to do? (add/remove/show/search/done): ")
print("OK,goodbye")
```
